```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Sample data
X = np.array([1, 2, 3, 4, 5])
y = np.array([1, 1.3333, 1.5, 1.6, 1.6667])
```

```python
# Define the nonlinear function
def nonlinear_model(X, a, b):
    return (a * X) / (b + X)
```

```python
params, _ = curve_fit(nonlinear_model, X, y)

a, b = params
```

```python
# Predictions
y_pred = nonlinear_model(X, a, b)
```

```python
# Plot
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title('Nonlinear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Print fitted parameters
print(f"Fitted parameters: a={a:.3f}, b={b:.3f}")
```
